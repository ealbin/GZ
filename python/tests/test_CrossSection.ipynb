# Plot Cross Sections

```python
e = np.linspace(0, 1000, 1001)*1e6 # energy axis [eV] 

He = []
C  = []
O  = []
Fe = []
U  = []

for energy in e:
    He.append( CrossSection.singleNucleon(4, energy)/1e-27 ) # cross section [mb]

for energy in e:
    C.append( CrossSection.singleNucleon(12., energy)/1e-27 )

for energy in e:
    O.append( CrossSection.singleNucleon(16., energy)/1e-27 )

for energy in e:
    Fe.append( CrossSection.singleNucleon(56., energy)/1e-27 )

for energy in e:
    U.append( CrossSection.singleNucleon(238., energy)/1e-27 )

plt.figure(figsize=(10,7))

He_line, = plt.loglog(e/1e6, He, 'm', label='Helium-4')
C_line,  = plt.loglog(e/1e6, C,  'k', label='Carbon-12')
O_line,  = plt.loglog(e/1e6, O,  'b', label='Oxygen-16')
Fe_line, = plt.loglog(e/1e6, Fe, 'r', label='Iron-56')
U_line,  = plt.loglog(e/1e6, U,  'g', label='Uranium-238')

plt.grid(True)
plt.title('Single Nucleon Photoejection Cross Section')
plt.xlabel('Photon Energy [MeV]')
plt.ylabel('Cross Section [mb]')
plt.legend()

plt.show()
```
